```python
import pandas as pd
import os
import FinanceDataReader as fdr

# CSV 파일 경로 설정
file_path = "./kopsi_200_stocks.csv"

# CSV 파일 읽어오기
df = pd.read_csv(file_path)

# code 열에서 중복값 제거하여 고유한 종목 코드 가져오기
unique_codes = df['code'].unique()
# "stock_data" 폴더 경로 설정
folder_path = "stock_data"

# "stock_data" 폴더 생성
os.makedirs(folder_path, exist_ok=True)

# 각 종목 코드에 대해 주가 데이터 수집
for code in unique_codes:
    # 종목 코드에서 작은따옴표 제거
    code = code.replace("'", "")
    
    try:
        # 주가 데이터 수집
        stock_data = fdr.DataReader(code, start='2021-01-01', end='2024-02-29')
        
        # 주가 데이터를 CSV 파일로 저장
        file_name = f"{code}.csv"
        file_path = os.path.join(folder_path, file_name)
        stock_data.to_csv(file_path)
        print(f"{code} stock data saved to {file_path}")
    except Exception as e:
        print(f"Failed to fetch stock data for {code}: {str(e)}")

print("Stock data collection completed.")
```
